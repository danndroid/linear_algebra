```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize
```

```python
def get_field(A):
    x, y ,z = A[0][0], A[1][0], A[1][1]
    
    return x,y,z

def get_str(A):
    x, y ,z = A[0][0], np.round(A[1][0],2), A[1][1]
    
    return '-'.join([str(s) for s in [x,y,z]])

def surface_points(size=10, limit=1):
    x = np.linspace(0, limit, size)
    z = np.linspace(0, limit, size)
    
    xs = []
    zs = []
    yas = []
    ybs = []
    for i in x:
        for j in z:
            a = np.sqrt(i*j)
            b = -1*np.sqrt(i*j)

            xs.append(i)
            zs.append(j)
            yas.append(a)
            ybs.append(b)
            
    return xs,yas,ybs,zs
```

```python
def genrate_matrix(r=None):
    if r == None:
        r = np.random.rand()
        s = np.random.binomial(1, 0.5)
        if s:
            r = -r
    
    I = np.array([[1,0],
                 [0,1]])
    B = np.array([[0,1],
                 [1,0]]) 
    M = I + (B*r)
    
    return M
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.plot_trisurf(xs, yas, zs, linewidth=0.2, antialiased=True, color='gray', alpha=0.1)
ax.plot_trisurf(xs, ybs, zs, linewidth=0.2, antialiased=True, color='gray', alpha=0.5)
ax.scatter(*get_field(I), color='r', s=70, label=f'I={get_str(I)}')

for i in range(2):
    M = genrate_matrix()
    ax.scatter(*get_field(M), color='b')
    
    for t in np.linspace(0.1, 0.5, 5):
        ax.scatter(*get_field(M+t), color='b')

ax.invert_xaxis()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()            
plt.show()
```
