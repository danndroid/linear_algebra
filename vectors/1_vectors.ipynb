```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# ROW VECTORS
basis = np.array([[1, 0, 0],
                  [0, 1, 0],
                  [0, 0, 1]])
basis[0]
```

```python
origin = np.array([0,0,0])
origin
```

```python
def vec_line(u,v): 
    line = [[a,b] for a, b in zip(u,v)]
    return line
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')
ax.set_title('Canonical vectors in R3', fontsize=15)

l = 1
l_axis = np.arange(-l,l+1)
ax.view_init(15, 25, 'z')

for i in l_axis:
    for j in l_axis:
        for k in l_axis:
            ax.scatter(i,j,k, color='gray', marker='.', s=25)
ax.scatter(*origin, color='crimson', marker='+', s=100) # grid

#ax.plot([0,0],[0,1],[1,0], '--', color='gray') # (u1,v1), (u2,v2), (u3,v3)  
#ax.plot3D(*vec_line(basis[0], basis[2]), '--', color='blue') # (u1,v1), (u2,v2), (u3,v3)

#ax.quiver(0, 0, 0, basis[:,0], basis[:,1], basis[:,2], color='steelblue')
plt.quiver(0,0,0, *basis[0], color='black', label=f'{basis[0]}') # X
plt.quiver(0,0,0, *basis[1], color='blue', label=f'{basis[1]}') # Y
plt.quiver(0,0,0, *basis[2], color='green', label=f'{basis[2]}') # Z


ax.set_xlim([-l, l])
ax.set_ylim([-l, l])
ax.set_zlim([-l, l])

ax.set_xticks(l_axis)
ax.set_yticks(l_axis)
ax.set_zticks(l_axis)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.set_title('Vector addition', fontsize=15)

l = 1
l_axis = np.arange(-l,l+1)
ax.view_init(15, 25, 'z')

for i in l_axis:
    for j in l_axis:
        for k in l_axis:
            ax.scatter(i,j,k, color='gray', marker='.', s=25)
ax.scatter(*origin, color='crimson', marker='+', s=100) # grid

#for v in basis:
#    print(v)
#    plt.quiver(0,0,0, *v, label=f'{v}')
plt.quiver(0,0,0, *basis[0], color='black', label=f'{basis[0]}') # X
plt.quiver(0,0,0, *basis[1], color='blue', label=f'{basis[1]}') # Y
plt.quiver(0,0,0, *basis[2], color='green', label=f'{basis[2]}') # Z

ax.quiver(0, 0, 0, *(basis[1]+basis[2]), ls='dashed', color='gray', label=f'{basis[1]}+{basis[2]}={basis[1]+basis[2]}')

ax.set_xlim([-l, l])
ax.set_ylim([-l, l])
ax.set_zlim([-l, l])

ax.set_xticks(l_axis)
ax.set_yticks(l_axis)
ax.set_zticks(l_axis)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.legend()
plt.tight_layout()
plt.show()
```
